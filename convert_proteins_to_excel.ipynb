```python
import pandas as pd
```

```python
s = ""
with open("IBP query amino acid sequence list.fasta") as f:
  s = f.read()
```

```python
a = s.split(">")[1:]

```

```python
lists = []
for protein_sequence in a: 
    temp = dict()
    # print(protein_sequence.partition(" "))
    temp["id"], _, after = protein_sequence.partition(" ")
    # print(temp["id"])
    temp["notes"], _, after = after.partition("[")
    temp["species_name"], _, after = after.partition("]")
    temp["code"] = after.strip("\n")

    lists.append(temp)

df = pd.DataFrame(lists)
```
