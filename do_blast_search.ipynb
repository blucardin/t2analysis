```python
from Bio import Blast, SeqIO
```

```python
fasta_file_name = "IBP query amino acid sequence list.fasta"

with open(fasta_file_name) as fasta_file:
    fasta_string = fasta_file.read()
```

```python
records = list(SeqIO.parse("IBP query amino acid sequence list.fasta", "fasta"))
```

```python
records
```

txid10197 is for Ctenophores

```python
result_stream = Blast.qblast("blastp", "nr_cluster_seq", fasta_string, expect=0.05, word_size = 5, entrez_query="txid10197[ORGN]")
with open("blastp_results.xml", "wb") as out_stream:
    out_stream.write(result_stream.read())
```

The results from the web page blast are identical to the result stream.

(except the query ids of course, they change for every query)

Nevertheless, I used the results from the web just to be safe.

```python
result_stream = Blast.qblast("tblastn", "core_nt", fasta_string, expect=0.05, word_size=5,
                             entrez_query="txid10197[ORGN]")

with open("tblastn_results.xml", "wb") as out_stream:
    out_stream.write(result_stream.read())
```
